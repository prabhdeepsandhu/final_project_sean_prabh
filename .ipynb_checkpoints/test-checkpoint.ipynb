```python
df['artists'] = df['artists'].str.strip('[]').str.replace(' ','').str.replace("'",'')
df['artists'] = df['artists'].str.split(',')
```

```python
artistList = []
for index, row in df.iterrows():
    artists = row['artists']
    
    for artist in artists:
        if artist not in artistList:
            artistList.append(artist) 
```

```python
def binary(artist_list):
    binaryList1 = []
    
    for artist in artistList:
        if artist in artist_list:
            binaryList1.append(1)
        else:
            binaryList1.append(0)
    
    return binaryList1
```

```python
df['artist_bin'] = df['artists'].apply(lambda x: binary(x))
df['artist_bin'].sample(5)
```

```python
from scipy import spatial

def Similarity(song1, song2):
    a = df.iloc[song1]
    b = df.iloc[song2]
    
    artistA = a['artist_bin']
    artistB = b['artist_bin']
    artistDistance = spatial.distance.cosine(artistA, artistB)
    
    danceabilityA = a['danceability_bin']
    danceabilityB = b['danceability_bin']
    danceabilityDistance = spatial.distance.cosine(danceabilityA, danceabilityB)
    
    energyA = a['energy_bin']
    energyB = b['energy_bin']
    energyDistance = spatial.distance.cosine(energyA, energyB)
    
    speechinessA = a['speechiness_bin']
    speechinessB = b['speechiness_bin']
    speechinessDistance = spatial.distance.cosine(speechinessA, speechinessB)
        
    acousticnessA = a['acousticness_bin']
    acousticnessB = b['acousticness_bin']
    acousticnessDistance = spatial.distance.cosine(acousticnessA, acousticnessB)
    
    livenessA = a['liveness_bin']
    livenessB = b['liveness_bin']
    livenessDistance = spatial.distance.cosine(livenessA, livenessB)
    
    valenceA = a['valence_bin']
    valenceB = b['valence_bin']
    valenceDistance = spatial.distance.cosine(valenceA, valenceB)
    
    yearA = a['year_bin']
    yearB = b['year_bin']
    yearDistance = spatial.distance.cosine(yearA, yearB)
    
    tempoA = a['tempo_bin']
    tempoB = b['tempo_bin']
    tempoDistance = spatial.distance.cosine(tempoA, tempoB)
    
    return artistDistance + danceabilityDistance + energyDistance + speechinessDistance + acousticnessDistance + livenessDistance + valenceDistance + yearDistance + tempoDistance
```
